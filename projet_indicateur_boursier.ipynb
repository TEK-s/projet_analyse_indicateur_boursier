# I-3 visualisation via des indicateurs de trading

l'indicateur MACD (Moving Average Convergence Divergence) : c'est un indicateur utilisé pour repérer un signal d’achat ou de vente. il utilise 2 moving average dont une sur petite periode(ligne représentée en rouge) et une autre sur une plus grande période(ligne représentée en vert). Ainsi lorsque la ligne verte traverse à la hausse la ligne rouge, c’est considéré comme un signal d’achat. À l’opposé, c’est un signal de vente.


l'indicateur rsi(Relative Strength Index) : est généralement utilisé pour détecter les marchés surachetés et surventés.
- Si le RSI se situe autour du niveau 80, on suppose que l'instrument est sur-acheté. C'est une situation dans laquelle les prix ont augmenté de manière forte.
- Si le RSI est autour du niveau 20, c'est un signal technique traduisant que l'instrument peut être sur-vendu. Dans ce cas, c'est une situation dans laquelle les prix ont fortement chute.


l'indicateur bollinger : se base sur l'utilisation de l'écart type et du canal de volatilité pour déterminer où une bande de support ou de résistance probable qui pourrait prendre place.Il s'agit de tracer des lignes au-dessus et en-dessous d'une mesure centrale du prix. Ces lignes, également appelées enveloppes ou bandes, s'élargissent ou se contractent selon la volatilité ou l'absence de volatilité du marché concerné.

```python
def indicateur_macd(data,l1=12,l2=26,l3=9):
    MME_courte=data.ewm(span=l1, adjust=False).mean()
    MME_longue=data.ewm(span=l2, adjust=False).mean()
    macd=MME_courte-MME_longue
    signal=macd.ewm(span=l3, adjust=False).mean()
    return macd, signal

def indicateur_rsi(data,periode):
    différence=data.diff()
    valeur_hausse=différence.copy()
    valeur_baisse=différence.copy()
    valeur_hausse[valeur_hausse<0]=0
    valeur_baisse[valeur_hausse>0]=0
    moyenne_gain=valeur_hausse.ewm(com=periode, adjust=False).mean()
    moyenne_perte=valeur_baisse.ewm(com=periode, adjust=False).mean()
    Rs=moyenne_gain/(-1*moyenne_perte)
    RSI=(100 - (100 / (1 + Rs)))
    return RSI

def indicateur_bollinger(data,periode):
    SMA=data.rolling(periode).mean()
    HAUT_SMA=np.array(SMA)+(2*np.array(data.rolling(periode).std()))
    BAS_SMA=np.array(SMA)+(-2*np.array(data.rolling(periode).std()))
    return HAUT_SMA,BAS_SMA
```

# II- ENTRAINEMENT ET PREDICTION

Dans le but de préparer l'entrainement de la machine la fonction valeur_classe est définie afin de determiner la tendance de l'action en fonction d'un nombre de jour. ainsi on compare de manière successive un jour et son jour suivant, afin de compter le nombre de fois où on observe une augmentation et le nombre de fois où une baisse est observée pendant cette période.
De cette façon si le nombre d'élément en hausse est plus grand on conclu que la tendance est haussière ainsi sur cette période dans un tableau on stocke une suite de 1 sinon on stocke une suite de 0.

```python
def valeur_classe(data,nbr_jour):
    max_values=data.shape[0]
    i=0
    n=0
    hausse=0
    baisse=0
    tab=[]
    while(n!=1):
        for j in range(nbr_jour):
            if(i+j+2<=max_values): 
                if (history_lvmh[i+j:i+j+1]['200MVA'].values[0]>history_lvmh[i+j+1:i+j+2]['200MVA'].values[0]):
                    baisse=baisse+1
                else:
                    hausse=hausse+1
            else:
                n=1
                break
        if(n!=1):        
            if (hausse>baisse):
                for j in range(i,nbr_jour+i):
                    tab.append(1)
            else:
                for j in range(i,nbr_jour+i):
                    tab.append(0)
            hausse=0
            baisse=0
            i=i+nbr_jour
        else:
            if (hausse>baisse):
                for j in range(i,j+i):
                    tab.append(1)
            else:
                for j in range(i,j+i):
                    tab.append(0)
    return tab

```

- pour la prédiction le choix du random forest c'est fait parce qu'il n'y a pas de relation linéaire entre les attributs et la classe que l'on souhaite prédire.
- le backtesting est une méthode permettant d'évaluer la viabilité d’une stratégie de trading en découvrant comment elle fonctionnerait en utilisant des données historiques. Ainsi la fonction backtesting permet d'entrainer le modèle de la première date disponible dans le jeu de donnée jusqu'a une date fixée(le paramètre debut) puis de faire le test avec les données entre  la date fixé et a date fixé + la step. Ainsi on évalue de manière successive en entrainant par la suite la machine avec les donnée entre la toute première date et la date de debut + step puis on effectue les test avec les données comprises entre la date de debut + step et la date de debut + step + step et ainsi de suite
- à la fin de chaque test les résultats sont conservés.

```python
def prediction(x_train, y_train,x_test, y_test,index_data,model):
    model.fit(x_train, y_train)
    predictions= model.predict(x_test)
    predictions=pd.Series(predictions, index=index_data, name='prediction')
    resultats_combine=pd.concat([y_test,predictions],axis=1)
    return resultats_combine


def backtesting(x,y,data, model,debut,step):
    predictions=[]
    max_values=x.shape[0]
    
    for i in range(debut,max_values,step):
        x_train=x[:i]
        y_train=y[:i]
        x_test=x[i:i+step]
        y_test=y[i:i+step]
        index_data=data[i:i+step].index
        
        resultat=prediction(x_train, y_train,x_test, y_test,index_data,model)
        predictions.append(resultat)
    return pd.concat(predictions)
```

dans notre cas de figure l'entrainement de la machine est fait de la première date du jeu de donnée jusqu'au 2020-06 et la serie de test c'est faite avec un pas de 25 jours

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestClassifier

model=RandomForestClassifier(n_estimators=100,min_samples_split=2,random_state=1)
y=history_lvmh['class']
x=history_lvmh.drop(['class'],axis=1).values

resultat=backtesting(x,y,history_lvmh,model,3750,25)
```

pour évaluer les performances de la prédiction on calcul les gains obtenu en considérant que l'on mise sur une tendance haussière.
Si par exemple on souhaite miser à chaque fois uniquement sur une tendance haussière au cours des périodes de 25 jours, lorsque le nombre de 1 prédit sur cette période est supérieur au nombre de 0 alors à ce moment on considère que l'on peut calculer le gain car la tendance est haute.
- si le gain est positive ça suppose que la machine a fait une bonne prédiction donc effectivement la tendance est haussière
- si le gain est négatif ça suppose qu'en réalité la tendance est baissière mais la machine a prédit le contraire

Profit/perte = (Prixt1 – Prixt0) x N x FXt1

Où :

- Prixt0 = prix par unité à l’ouverture du trade
- Prixt1 = prix par unité lors de la visualisation ou de la fermeture d’une position
- n = nombre d’unités
- FX(t1) = taux de change entre la devise de base de l’actif et l’USD au moment où la position est ouverte pour la première fois

```python
def retour_investissement(resultat,data,n,change):
    #pip=(0.01/143.35)*mise*0.5*100*1.36
    total=0
    classe=resultat['class'].tolist()
    prediction=resultat['prediction'].tolist()
    gain_mois=[]
    for i in range(0,len(resultat),25):
        if (prediction[i:i+25].count(1)>prediction[i:i+25].count(0)):
            total=total+(n*(data['open'][i]-data['close'][i])*change)
        a=pd.to_datetime(data[i:i+1].index.values[0])
        total=f"{a.year}-{a.month} le gain vaut : {float(total)}"
        gain_mois.append(total)
        total=0
    display(gain_mois)
```
